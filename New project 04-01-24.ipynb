In this column there are some outliers 

For the removing outliers i will use IQR method for Change column

#### IQR for remove outliers 

```python
q1 = new_df["change"].quantile(0.25)
q3 = new_df["change"].quantile(0.75)
iqr = q3-q1
```

```python
q1 ,q3 , iqr
```

```python
upper_limit = q3 + (1.5* iqr)
lower_limit = q1 - (1.5* iqr)
```

```python
lower_limit ,upper_limit
```

```python
sns.boxplot(new_df["change"])
```

```python
new_df.loc[(new_df["change"] > upper_limit)|  (new_df["change"] < lower_limit)]
```

```python
#trimming (DELETE THE OUTLIERS)
df1 = new_df.loc[(new_df["change"]< upper_limit) &  (new_df["change"]> lower_limit)]
print("before removing outliers :", len(new_df))
print("after  removing outliers :", len(df1))
print("outliers :" , len(new_df) - len(df1))
```

```python
sns.boxplot(df1["change"])
```

Now this column is without outlier

Now in another column we will use zscore method for removing outlies

```python
sns.distplot(new_df["change_percent"])
```

```python
sns.boxplot(new_df["change_percent"])
```

we can see clearly in this column some are outliers

```python
upper_limit = new_df["change_percent"].mean() + 3*new_df["change_percent"].std()
lower_limit = new_df["change_percent"].mean() - 3*new_df["change_percent"].std()
print("upper_limit:", upper_limit)
print("lower_limit:", lower_limit)
```

```python
new_df.loc[(new_df["change_percent"] > upper_limit) | (new_df["change_percent"] < lower_limit)]
```

```python
# in this column we will capping the outliers 
df1.copy()
df1.loc[df1["change_percent"]> upper_limit , "change_percent"] = upper_limit
df1.loc[df1["change_percent"]< lower_limit , "change_percent"] = lower_limit
```

```python
sns.boxplot(df1["change_percent"])
```

in this menthod we were not trim data for delete data, we just cap it
